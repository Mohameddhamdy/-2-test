```python
import pandas as pd

df = pd.read_csv("/kaggle/input/home-credit-default-risk/application_test.csv")
df
```

```python
#To determine which features are available in the dataset, you can simply print the column names:
features = df.columns.tolist()
features
```

```python
#To identify the categorical features and their types (nominal, ordinal, ratio, or interval-based), you can iterate through the columns and check the unique values:
categorical_features = []

for feature in df.columns:
 unique_values = df[feature].nunique()

 if unique_values <= 5:
  categorical_features.append((feature, 'Nominal')) 
 else:
  categorical_features.append((feature, 'Ordinal'))
print(categorical_features)
```

```python
#To identify the numerical features and their types (discrete, continuous, or timeseries-based), you canuse the following code:
numerical_features = []

for feature in df.columns:
  if df[feature].dtype in ['float64']:
    numerical_features.append((feature, 'Continuous')) 

print(numerical_features)
```

```python
#To identify features that contain mixed data types (numerical and alphanumeric data within the samefeature), you can iterate over the columns and check the data type of each value within the column. If there are multiple data types present, it indicates mixed data types.
mixed_features = []

for feature in df.columns:
 data_types = df[feature].apply(type)
 unique_data_types = data_types.unique()

 if len(unique_data_types) > 1:
  mixed_features.append(feature)

print(mixed_features)
```

```python
#To identify features that contain blank, null, or empty values, you can use the `isnull` function along with the `sum` function to get the count of missing values in each feature.
missing_values = df.isnull().sum()
missing_features = missing_values[missing_values > 0]

print(missing_features)
```

```python
#To assess how well the sample (train dataset) represents the actual problem domain, you can compare the proportion of deaths in the train dataset to the overall proportion of deaths in the total population.
total_population = 2224
total_deaths = 1502
train_deaths = df["AMT_INCOME_TOTAL"].sum()

train_proportion = train_deaths / len(df)
actual_proportion = total_deaths / total_population

if train_proportion > actual_proportion:
 print("The train dataset over-represents deaths.")
elif train_proportion < actual_proportion:
 print("The train dataset under-represents deaths.")
else:
 print("The train dataset represents the proportion of deaths accurately.")
```
